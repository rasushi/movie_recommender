```python
import pandas as pd
import numpy as np
import ast
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies_m = movies_merged[['movie_id','title','genres','keywords','overview','release_date','cast','crew']] #selecting useful columns
```

```python
movies_m.head()
```

```python
movies_m.isnull().sum()
```

```python
movies_m = movies_m.dropna()
movies_m.isnull().sum()
```

```python
movies_m.duplicated().sum()
```

```python
def extractor(obj):#to extract genre names and keyword names
    names = []
    for dict in ast.literal_eval(obj):
        names.append(dict['name'])
    return names
    
```

```python
movies_m['genres'] = movies_m['genres'].apply(extractor)
```

```python
movies_m['keywords'] = movies_m['keywords'].apply(extractor)
```

```python
movies_m.head(2)
movies_m.isnull().sum()
```

```python
def extract_3_cast(obj):#to extract genre names and keyword names
    names = []
    count = 0
    for dict in ast.literal_eval(obj):
        if count < 3:
            names.append(dict['name'])
            count += 1
        else:
            break
    return names
```

```python
movies_m['cast'] = movies_m['cast'].apply(extract_3_cast)
```

```python
movies_m.head(2)
movies_m.isnull().sum()
```

```python
def extract_director(obj):#to extract genre names and keyword names
    names = []
    for dict in ast.literal_eval(obj):
        if dict['job'] == 'Director':
            names.append(dict['name'])
            break
    return names
```

```python
movies_m['crew'] = movies_m['crew'].apply(extract_director)
```

```python
movies_m.head(2)
```

```python
movies_m['overview'] = movies_m['overview'].apply(lambda x:x.split())
```

```python
movies_m['release_year'] = (pd.to_datetime(movies_m['release_date'], errors='coerce')).dt.year
```

```python
movies_m['release_year'] = movies_m['release_year'].apply(lambda x: [x] if pd.notnull(x) else [])
movies_m.drop('release_date', axis = 1, inplace = True)
```

```python
movies_m.head(2)
```

```python
#removing spaces bw names to avoid confusion
movies_m['genres'] = movies_m['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies_m['keywords'] = movies_m['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies_m['crew'] = movies_m['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies_m['cast'] = movies_m['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_m.head(5)
```

```python
movies_m['tags'] = movies_m['overview'] + movies_m['genres'] + movies_m['keywords'] + movies_m['cast'] + movies_m['crew'] + movies_m['release_year']
```

```python
movies_m.head(5)
```

```python
new_movies = movies_m[['movie_id','title','tags']]
```

```python
new_movies.head(5)
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x:" ".join(str(i) for i in x)) #converting tags into a string by joining,str(i) was used to deal with int
```
